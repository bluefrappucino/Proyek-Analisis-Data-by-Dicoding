## Visualization & Explanatory Analysis

```python
import plotly.express as px
```

1. Berapa persentase peningkatan jumlah sewa pada musim panas dibandingkan musim lainnya?

```python
# Hitung rata-rata jumlah sewa per musim
df_season_avg = df_day.groupby("season")["cnt"].mean()

# Hitung persentase peningkatan musim panas
summer_avg = df_season_avg[2]
other_seasons_avg = df_season_avg.drop(2).mean()
increase_pct = ((summer_avg - other_seasons_avg) / other_seasons_avg) * 100

# Cetak hasil
print(f"Peningkatan jumlah sewa pada musim panas: {increase_pct:.2f}%")

```

2. Di jam berapa rata-rata jumlah sewa mencapai puncaknya?

```python
# Hitung rata-rata jumlah sewa per jam
df_hourly = df_hour.groupby("hr")["cnt"].mean()

# Cari jam dengan rata-rata tertinggi
peak_hour = df_hourly.idxmax()

# Cetak hasil
print(f"Jam dengan rata-rata jumlah sewa tertinggi: {peak_hour}")

# Membuat dataframe dari rata-rata jumlah sewa per jam
df_hourly = df_hour.groupby("hr")["cnt"].mean().reset_index()

# Plot
fig = px.line(df_hourly, x='hr', y='cnt', title='Rata-rata Jumlah Sewa per Jam')
fig.update_layout(xaxis_title='Jam', yaxis_title='Rata-rata Jumlah Sewa')
fig.show()
```

3. Apakah terdapat perbedaan signifikan dalam jumlah sewa pada hari kerja dan akhir pekan?

```python
import scipy.stats as stats

# Hitung rata-rata jumlah sewa pada hari kerja dan akhir pekan
df_weekday_avg = df_day.groupby("weekday")["cnt"].mean()

# Lakukan uji t-test untuk membandingkan rata-rata
t_stat, p_value = stats.ttest_ind(df_day[df_day["weekday"] <= 4]["cnt"], df_day[df_day["weekday"] >= 5]["cnt"])

# Interpretasi hasil
if p_value < 0.05:
    print("Terdapat perbedaan signifikan dalam jumlah sewa pada hari kerja dan akhir pekan (p-value < 0.05).")
else:
    print("Tidak terdapat perbedaan signifikan dalam jumlah sewa pada hari kerja dan akhir pekan (p-value >= 0.05).")

# Hitung rata-rata jumlah sewa per hari
df_weekday_avg = df_day.groupby("weekday")["cnt"].mean().reset_index()
df_weekday_avg['weekday'] = df_weekday_avg['weekday'].map({0: 'Minggu', 1: 'Senin', 2: 'Selasa', 3: 'Rabu', 4: 'Kamis', 5: 'Jumat', 6: 'Sabtu'})

# Buat bar plot
fig = px.bar(df_weekday_avg, x='weekday', y='cnt', title='Rata-Rata Jumlah Sewa per Hari', labels={'weekday': 'Hari', 'cnt': 'Rata-Rata Jumlah Sewa'})
fig.update_traces(marker_color='skyblue')
fig.update_layout(xaxis={'categoryorder': 'array', 'categoryarray': ['Minggu', 'Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu']},
                  yaxis_title='Rata-Rata Jumlah Sewa', xaxis_title='Hari', showlegend=False)
fig.show()
```

4. Berapa banyak peningkatan jumlah sewa yang diprediksikan jika temperatur naik 1 derajat Celcius?

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Membagi data menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(df_day[["temp"]], df_day["cnt"], test_size=0.25)

# Melatih model regresi linear
model = LinearRegression()
model.fit(X_train, y_train)

# Mengevaluasi model
score = model.score(X_test, y_test)
print(f"R-squared: {score:.2f}")

# Memprediksi jumlah sewa pada temperatur 1 derajat Celcius lebih tinggi
predicted_cnt = model.predict([[1]])

# Menghitung peningkatan jumlah sewa
increase_cnt = predicted_cnt[0] - y_test.mean()

# Mencetak hasil
print(f"Peningkatan jumlah sewa diprediksikan: {increase_cnt:.2f}")

# Visualisasi data
fig = px.scatter(x=X_test["temp"], y=y_test, title="Scatter Plot: Suhu vs Jumlah Sewa", labels={"temp": "Suhu (Celcius)", "cnt": "Jumlah Sewa"})
fig.add_scatter(x=[1], y=[predicted_cnt[0]], mode='markers', name='Prediksi', marker=dict(color='red', size=10, symbol='cross'))
fig.show()
```

```python
# Membuat scatter plot untuk data uji
fig = px.scatter(x=X_test.squeeze(), y=y_test, title='Regresi Linear: Pengaruh Suhu terhadap Jumlah Sewa', labels={'x': 'Suhu (Celcius)', 'y': 'Jumlah Sewa'}, trendline="ols")
fig.update_traces(marker=dict(color='blue'), mode='markers', name='Data Uji')

# Menambahkan garis regresi
fig.add_scatter(x=X_test.squeeze(), y=model.predict(X_test), mode='lines', line=dict(color='red', width=2), name='Regresi Linear')

# Menampilkan legenda
fig.update_layout(showlegend=True)

# Menampilkan plot
fig.show()
```

5. Rata-rata jumlah sewa per jam dihitung dengan menjumlahkan total sewa di setiap jam dan membaginya dengan total hari dalam dataset.

```python
# Hitung rata-rata jumlah sewa per jam
df_hourly = df_hour.groupby("hr")["cnt"].mean().reset_index()

# Plot
fig = px.line(df_hourly, x='hr', y='cnt', markers=True, title='Rata-rata Jumlah Sewa per Jam')
fig.update_layout(xaxis_title='Jam', yaxis_title='Rata-rata Jumlah Sewa', xaxis=dict(tickmode='linear'))
fig.show()
```

6. Peningkatan jumlah sewa yang diprediksikan dihitung dengan menggunakan model regresi linear.

```python
# Membagi data menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(df_day[["temp"]], df_day["cnt"], test_size=0.25)

# Melatih model regresi linear
model = LinearRegression()
model.fit(X_train, y_train)

# Mengevaluasi model
score = model.score(X_test, y_test)
print(f"R-squared: {score:.2f}")

# Memprediksi jumlah sewa pada temperatur 1 derajat Celcius lebih tinggi
predicted_cnt = model.predict([[1]])

# Menghitung peningkatan jumlah sewa
increase_cnt = float(predicted_cnt[0])

# Mencetak hasil
print(f"Peningkatan jumlah sewa diprediksikan: {increase_cnt:.2f}")
```

```python
# Membuat dataframe untuk prediksi suhu
df_pred = pd.DataFrame({'temp': [1], 'predicted_cnt': predicted_cnt})

# Plot
fig = px.scatter(df_pred, x='temp', y='predicted_cnt', title='Prediksi Jumlah Sewa pada Temperatur 1 derajat Celcius Lebih Tinggi')
fig.update_layout(xaxis_title='Temperatur (Celcius)', yaxis_title='Prediksi Jumlah Sewa')
fig.show()
```
